```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the grid world
grid_size = 10
world_map = np.zeros((grid_size, grid_size))

# Place the agent and red flag
agent_position = (0, 0)
red_flag_position = (9, 9)
world_map[agent_position] = 1  # Agent represented by 1
world_map[red_flag_position] = 2  # Red flag represented by 2

# Place the mines
mines_positions = [(4, 5), (4, 6), (4, 7), (5, 2), (5, 3), (5, 4)]
for mine_position in mines_positions:
    world_map[mine_position] = -1  # Mines represented by -1

# Visualize the grid world
plt.imshow(world_map, cmap='viridis', origin='lower', extent=[0, grid_size, 0, grid_size])

# Add grid lines
# plt.grid(True, color='black', linewidth=1)

# Add labels to cells
# for i in range(grid_size):
#     for j in range(grid_size):
#         plt.text(j + 0.5, i + 0.5, f'({j}, {i})', ha='center', va='center')

plt.title('Grid World with Agent, Red Flag, and Mines')
plt.show()

```
